```python
import matplotlib.pyplot as plt
plt.rc('font',size=23) #set default font size
plt.plot([1,2,3,4],lw=5,label="with apples") #lw = line width, label = legend text
plt.xlabel('minions') #set x label
plt.ylabel('fun  [laughts/second]') #set y label
plt.xlim ( (0,4) ) #set x range
plt.legend() #show legend
```
